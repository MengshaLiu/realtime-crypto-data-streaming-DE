```python
# It seems the data from the document spans over several periods. Let's first generate a chart to illustrate Tokyo house and apartment price trends over the past 10 years.

import matplotlib.pyplot as plt
import numpy as np

# Sample data (assumed from snippets, adjust accordingly)
years = np.arange(2018, 2024)

# House price index and apartment price index data (dummy values for illustration)
house_price_index = [3403, 3404, 3396, 3722, 4035, 4102 ]# House price index from 2018 to 2023
apartment_price_index = [3333, 3439, 3573, 3872, 4278, 4555]  # Apartment price index from 2018 to 2023

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(years, house_price_index, label='House Price Index', marker='o', linestyle='-', color='blue')
plt.plot(years, apartment_price_index, label='Apartment Price Index', marker='o', linestyle='-', color='green')

plt.title("Tokyo House and Apartment Price Trends (2014-2024)", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Price Index", fontsize=12)
plt.legend()
plt.grid(True)
```
